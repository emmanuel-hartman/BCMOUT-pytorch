```python
import os
os.chdir('..')
import matplotlib.pyplot as plt
import torch
from bcmout.Euclidean import Euclidean
Rn = Euclidean(3)
b = Rn.random(samples=10)
print(Rn)
print(Rn._metric)
print(torch.all(Rn.belongs(b)))
```

```python
d=Rn.dissimilarity(b)
plt.matshow(d.numpy())
plt.show()
print(d.shape)
```

```python
from sklearn.manifold import TSNE,MDS
X_embedded = MDS(n_components=2, dissimilarity='precomputed').fit_transform(d.numpy())
plt.scatter(X_embedded[:,0],X_embedded[:,1])
plt.scatter(X_embedded[:,0],X_embedded[:,1])
plt.show()
```

```python
from bcmout.ConeSpace import ConeOverM
COverS = ConeOverM(Rn,4)
print(COverS)
print(COverS._metric)
c=COverS.random(1000,10)
print(torch.all(COverS.belongs(c)))
```

```python
d=COverS.dissimilarity(c)
plt.matshow(d.numpy())
plt.show()
#print(d)
```

```python

X_embedded = MDS(n_components=2, dissimilarity='precomputed').fit_transform(d.numpy())
plt.scatter(X_embedded[:,0],X_embedded[:,1])
plt.scatter(X_embedded[:,0],X_embedded[:,1])
plt.show()
```

```python
from bcmout.MeasureSpace import MeasureSpace
MonS = MeasureSpace(Rn,4,metric="WFR")
print(MonS)
print(MonS._metric)
e = MonS.random(100,20,10)
print(torch.all(MonS.belongs(e)))
```

```python
d = MonS.dissimilarity(e,max_steps=10000,eps=.00001)
plt.matshow(d.numpy())
plt.show()
#print(d)
```

```python
X_embedded = MDS(n_components=2, dissimilarity='precomputed').fit_transform(d.numpy())
plt.scatter(X_embedded[:,0],X_embedded[:,1])
plt.scatter(X_embedded[:,0],X_embedded[:,1])
plt.show()
```
