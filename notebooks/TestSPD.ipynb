```python
import os
os.chdir('..')
import matplotlib.pyplot as plt
import torch
from bcmout.SPDMatrices import SPDMatrices
S = SPDMatrices(2)
b = S.random(samples=10,bound=1.0)
print(S)
print(S._metric)
print(torch.all(S.belongs(b)))
```

```python
d=S.dissimilarity(b)
plt.matshow(d.numpy())
plt.show()
print(d.shape)
```

```python
from sklearn.manifold import TSNE,MDS
X_embedded = MDS(n_components=2, dissimilarity='precomputed').fit_transform(d.numpy())
plt.scatter(X_embedded[:,0],X_embedded[:,1])
plt.scatter(X_embedded[:,0],X_embedded[:,1])
plt.show()
```

```python
from ConeSpace import ConeOverM, ExpMetric
metric= ExpMetric(S,2)
COverS = ConeOverM(S,2,metric=metric)
print(COverS)
print(COverS._metric)
c=COverS.random(100,10)
print(torch.all(COverS.belongs(c)))
```

```python
d=COverS.dissimilarity(c)
plt.matshow(d.numpy())
plt.show()
#print(d)
```

```python

X_embedded = MDS(n_components=2, dissimilarity='precomputed').fit_transform(d.numpy())
plt.scatter(X_embedded[:,0],X_embedded[:,1])
plt.scatter(X_embedded[:,0],X_embedded[:,1])
plt.show()
```

```python
from MeasureSpace import MeasureSpace
MonS = MeasureSpace(S,4)
print(MonS)
print(MonS._metric)
e = MonS.random(10,20,10)
print(torch.all(MonS.belongs(e)))
```
