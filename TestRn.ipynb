```python
import matplotlib.pyplot as plt
import torch
from Euclidean import Euclidean
Rn = Euclidean(3)
b = Rn.random(samples=1000)
print(Rn)
print(torch.all(Rn.belongs(b)))
```

```python
d=Rn.distance(b,b)
plt.matshow(d.numpy())
plt.show()
print(d.shape)
```

```python
from sklearn.manifold import TSNE,MDS
X_embedded = MDS(n_components=2, dissimilarity='precomputed').fit_transform(d.numpy())
plt.scatter(X_embedded[:,0],X_embedded[:,1])
plt.scatter(X_embedded[:,0],X_embedded[:,1])
plt.show()
```

```python
from ConeSpace import ConeOverM
COverS = ConeOverM(Rn,1)
print(COverS)
c=COverS.random(1000,10)
print(torch.all(COverS.belongs(c)))
```

```python
d=COverS.distance(c,c)
plt.matshow(d.numpy())
plt.show()
#print(d)
```

```python
from MeasureSpace import MeasureSpace
MonS = MeasureSpace(Rn,1)
print(MonS)
e = MonS.random(100,20,10)
print(torch.all(MonS.belongs(e)))
```
